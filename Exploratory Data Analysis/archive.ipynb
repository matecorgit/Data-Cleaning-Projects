```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data and see the first values
df = pd.read_csv('world_population.csv')
df.head()
```

```python
#Changing float format
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Info of the table
df.info()
```

```python
df.describe()
```

```python
#Checking the null values
df.isnull().sum()
```

```python
#Unique values
df.nunique()
```

```python
#Top 5 2022 countrys with the most population
df.sort_values(by='2022 Population', ascending=False).head()
```

```python
#Check the corralations
sns.heatmap(df.corr(), annot=True)
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

```python
#Grouping by continent with the most population growth
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population',ascending=False)
```

```python
df3 = df2.transpose()
```

```python
plt.plot(df3)
plt.xticks(rotation=45)
plt.show()
```
