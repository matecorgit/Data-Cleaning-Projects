```python
# Data Cleaning
```

```python
#Import libraries
import pandas as pd
```

```python
# Import data
df = pd.read_excel('Customer_Call_List.xlsx')
df.head()
```

```python
#Drop duplicates
df = df.drop_duplicates()
```

```python
df = df.drop(columns='Not_Useful_Column', axis=1)
```

```python
#Rename column
df = df.rename(columns={'Paying Customer' : 'Paying_Customer'})
```

```python
#Fixing Last Name
#df['Last_Name'] = df['Last_Name'].str.strip('/')
#df['Last_Name'] = df['Last_Name'].str.strip('...')
#df['Last_Name'] = df['Last_Name'].str.strip('_')
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
#Fixing phone number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
```

```python
#Fixing the data type pf phone number
df['Phone_Number'] = df['Phone_Number'].astype(str)
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
```

```python
#Deleting the NaNs
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
#Address
df['Street_Address'] = df['Address'].str.split(',').str[0]
df['State'] = df['Address'].str.split(',').str[1]
```

```python
#Deleting the old column of Address
df = df.drop(columns='Address')
```

```python
#Re-ordering columns
df = df[['CustomerID','First_Name','Last_Name','Phone_Number','Street_Address','State','Paying_Customer','Do_Not_Contact']]
```

```python
#Fixing Paying Customer
df['Paying_Customer'] = df['Paying_Customer'].str.replace('Yes','Y')
```

```python
df['Paying_Customer'] = df['Paying_Customer'].str.replace('Y','Yes')
```

```python
df['Paying_Customer'] = df['Paying_Customer'].str.replace('No','N')
```

```python
df['Paying_Customer'] = df['Paying_Customer'].str.replace('N','No')
```

```python
#Fixing Do_Not_Contact
df['Do_Not_Contact'] = df['Paying_Customer'].str.replace('Yes','Y')
```

```python
df['Do_Not_Contact'] = df['Paying_Customer'].str.replace('Y','Yes')
```

```python
df['Do_Not_Contact'] = df['Paying_Customer'].str.replace('Noo','N')
```

```python
df['Do_Not_Contact'] = df['Paying_Customer'].str.replace('N','No')
```

```python
#Deleting NaN
df = df.replace('N/a','')
df = df.fillna('')
```

```python
#More fixing
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Noo','No')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No/a','No')
df['Paying_Customer'] = df['Do_Not_Contact'].str.replace('No/a','No')
```
